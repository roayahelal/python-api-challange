```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
cities
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
cities_df = pd.DataFrame({"Random City": cities})
cities_df["Cloudiness"] = ""
cities_df["Country"] = ""
cities_df["Date"] = ""
cities_df["Humidity"] = ""
cities_df["lat"]  = ""
cities_df["lng"] = ""
cities_df["Max Temp"] = ""
cities_df["Wind Speed"] = ""

cities_df
```

```python
base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"

# Build partial query URL
query_url = f"{base_url}appid={api_key}&units={units}&q="

print("Beginning Data Retrieval")


for index, row in cities_df.iterrows():
    city_used = row["Random City"]
    
    try:
        print(f"Retrieving Results for Index {index}: {city_used}.")
        response = requests.get(query_url+city_used).json()
        cities_df["Cloudiness"].loc[cities_df["Random City"] == city_used] = response["clouds"]["all"]
        cities_df["Country"].loc[cities_df["Random City"] == city_used] = response["sys"]["country"]
        cities_df["Date"].loc[cities_df["Random City"] == city_used] =  response["dt"]
        cities_df["Humidity"].loc[cities_df["Random City"] == city_used] =  response["main"]["humidity"]
        cities_df["lat"].loc[cities_df["Random City"] == city_used]  =  response["coord"]["lat"]
        cities_df["lng"].loc[cities_df["Random City"] == city_used] =  response["coord"]["lon"]
        cities_df["Max Temp"].loc[cities_df["Random City"] == city_used] =  response["main"]["temp_max"]
        cities_df["Wind Speed"].loc[cities_df["Random City"] == city_used] =  response["wind"]["speed"]
    
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
    
    print("------------")
    time.sleep(1)
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
cities_df.to_csv("WeatherPy")
cities_df = pd.DataFrame(cities_df)
cities_df
```

### Plotting the Data


#### Latitude vs. Temperature Plot

```python

lat_temp = cities_df[["lat", "Max Temp"]]
lat_temp.replace('', np.nan, inplace=True)
lat_temp.dropna()
plt.scatter(lat_temp["lat"],lat_temp["Max Temp"], marker ="o", edgecolors="black",facecolors="pink")
lat_temp.count()
plt.title("City Latitude vs. Max Temperature")
plt.xlabel("Latitude")
plt.ylabel("Max Temperature (F)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("lat_MaxTemp.png")
plt.show()
```

#### Latitude vs. Humidity Plot

```python
lat_hu = cities_df[["lat", "Humidity"]]
lat_hu.replace('', np.nan, inplace=True)
lat_hu.dropna()
plt.scatter(lat_hu["lat"],lat_hu["Humidity"], marker ="o", edgecolors="black",facecolors="grey")
plt.title("City Latitude vs. Humidity")
plt.xlabel("Latitude")
plt.ylabel("Humidity")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("lat_humid.png")
plt.show()
lat_hu.count()
```

#### Latitude vs. Cloudiness Plot

```python
lat_cloud = cities_df[["lat", "Cloudiness"]]
lat_cloud.replace('', np.nan, inplace=True)
lat_cloud.dropna()
plt.scatter(lat_cloud["lat"],lat_cloud["Cloudiness"], marker ="o", edgecolors="black",facecolors="white")
plt.title("City Latitude vs. Cloudiness")
plt.xlabel("Latitude")
plt.ylabel("Cloudiness %")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("lat_cloud.png")
plt.show()
lat_cloud.count()
```

#### Latitude vs. Wind Speed Plot

```python
lat_wind = cities_df[["lat", "Wind Speed"]]
lat_wind.replace('', np.nan, inplace=True)
lat_wind.dropna()
plt.scatter(lat_wind["lat"],lat_wind["Wind Speed"], marker ="o", edgecolors="black",facecolors="blue")
plt.title("City Latitude vs. Wind Speed")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.grid (b=True,which="major",axis="both",linestyle="-",color="lightgrey")
plt.savefig("lat_cloud.png")
plt.show()
lat_wind.count()
```
